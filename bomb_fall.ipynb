```python
import c4dynamics as c4d
import numpy as np
```

```python
# This is a actual bomb
bomb = c4d.datapoint(x=0, z=50, vx=100, vz=0, mass=10, l=0.2, w=0.2, h=0.2)
```

```python
g = 9.8 # Gravitational constant
C = 1.05 # Drag Coefficient 
p = 1.2840 # Air density
```

```python
bomb_Sx = bomb.h * bomb.w
bomb_Sz = bomb.l * bomb.h
gravity_force = bomb.mass * g
```

```python
dt = 0.1
time = np.arange(0, 10, dt)

# Integrate a draw X and Z position charge over time
for t in time:
    Fx = -0.5 * C * p * bomb.vx ** 2 * bomb_Sx
    Fz = 0.5 * C * p * bomb.vz ** 2 * bomb_Sz - (bomb.mass * g) 
    Forces = [Fx, 0, Fz]
    bomb.inteqm(Forces, dt=dt)
    bomb.store(t)
    if (bomb.z) <= 0:
        break
bomb.draw("x")
bomb.draw("z")
```
